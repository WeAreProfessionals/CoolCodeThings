```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random points
n = 10 # Number of points
dim = 2  # Dimensions
# np.random.seed(2)
points = np.random.rand(n, dim)

plt.plot(points[:,0],points[:,1], 'ro')

# Calculate the centroid as the starting point
centroid = np.mean(points, axis=0)
# plt.plot(centroid[0],centroid[1],'ko')


# Define the cost function (sum of distances from a point)
def cost_function(point, points):
    distances = np.linalg.norm(points - point, axis=1)
    return np.sum(distances)

def get_gradient(point, points):
    x = 0
    y = 0
    for i in range(len(points)):
        x += (point[0]-points[i,0])/np.linalg.norm(points[i,:]-point)
        y += (point[1]-points[i,1])/np.linalg.norm(points[i,:]-point)
        
    return (x, y)

# Gradient Descent
learning_rate = 0.001
iterations = 1000

current_point = [1,1]
plt.plot(current_point[0],current_point[1],'ko')

for i in range(iterations):
    
    gradient = get_gradient(current_point, points)
    current_point[0] -= learning_rate * gradient[0]
    current_point[1] -= learning_rate * gradient[1]

    if i % 10 == 0:
        cost = cost_function(current_point, points)
        print(f"Iteration {i}: Cost = {cost:.4f}")

final_cost = cost_function(current_point, points)
print("Final point:", current_point)
print("Final cost:", final_cost)

plt.plot(current_point[0],current_point[1],'bo')

# [0.4999049168251894, 0.5023010914401557]
```
