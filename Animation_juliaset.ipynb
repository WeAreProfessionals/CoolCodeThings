```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
import cmath

# Define the parameters
width, height = 800, 800
xmin, xmax = -1.5, 1.5
ymin, ymax = -1.5, 1.5
a_values = np.linspace(0, 2 * np.pi, 200)

# Create a function to generate Julia Set for a given 'a'
def generate_julia_set(a):
    img = np.empty((width, height))
    for x in range(width):
        for y in range(height):
            zx, zy = x * (xmax - xmin) / (width - 1) + xmin, y * (ymax - ymin) / (height - 1) + ymin
            c = 0.355 * cmath.exp(1j * a)
            z = zx + zy * 1j
            for i in range(256):
                if abs(z) > 2.0:
                    break
                z = z * z + c
            img[x, y] = i
    return img

# Create the figure and axis
fig, ax = plt.subplots()

# Function to update the animation frame
def update(frame):
    ax.clear()
    a = a_values[frame]
    img = generate_julia_set(a)
    ax.imshow(img.T, origin='lower', extent=(xmin, xmax, ymin, ymax), cmap='twilight')

# Create the animation
ani = FuncAnimation(fig, update, frames=len(a_values), interval=50)

# Save the animation as a GIF
ani.save('julia_sets_animation_2.gif', writer='pillow')

# Show the animation
plt.show()
   
```
